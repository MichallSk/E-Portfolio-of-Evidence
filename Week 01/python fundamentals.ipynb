### Exercise 3: Conditions
Use simple `if`, `elif`, and `else` blocks to react to user input.

```python
"""Exercise 3.1: Classify a number as positive, negative, or zero."""
typed_number = input("Enter a number: ")
try:
    number_value = float(typed_number)
    if number_value > 0:
        print("The number is positive.")
    elif number_value < 0:
        print("The number is negative.")
    else:
        print("The number is zero.")
except ValueError:
    print("That was not a valid number.")
```

#### Exercise 3.2: Vowel or Consonant
Check the first letter of a word entered by the user.

```python
"""Exercise 3.2: Identify whether the first letter is a vowel or consonant."""
text_input = input("Enter a word: ").strip()
if text_input:
    first_char = text_input[0].lower()
    vowels = {"a", "e", "i", "o", "u"}
    if first_char in vowels:
        print(f"'{text_input}' starts with a vowel.")
    else:
        print(f"'{text_input}' starts with a consonant.")
else:
    print("No word entered to check for vowels.")
```

### Exercise 4: Loops
Use `for` and `while` loops for simple number sequences.

```python
"""Exercise 4: Practice loops."""
print("Numbers from 1 to 10:")
for number in range(1, 11):
    print(number, end=" ")
print("\n")

print("Even numbers from 2 to 10:")
for even in range(2, 11, 2):
    print(even, end=" ")
print("\n")

print("First 5 Fibonacci numbers:")
# Start the sequence with the first two numbers.
fib_numbers = [0, 1]
while len(fib_numbers) < 5:
    fib_numbers.append(fib_numbers[-1] + fib_numbers[-2])
print(*fib_numbers)
```

### Exercise 5: Functions
Define and call small helper functions to reuse logic.

```python
"""Exercise 5: Define and test simple functions."""
def add_numbers(a, b):
    """Return the sum of two numbers."""
    return a + b

def largest_number(values):
    """Return the largest number in a non-empty list."""
    return max(values)

def count_vowels(text):
    """Count vowels in the provided text."""
    vowels = {"a", "e", "i", "o", "u"}
    return sum(1 for char in text.lower() if char in vowels)

print("Sum of 7 and 5:", add_numbers(7, 5))
print("Largest number in [3, 9, 2, 11]:", largest_number([3, 9, 2, 11]))
print("Vowel count in 'Security':", count_vowels("Security"))
```

### Exercise 6: Password Strength Checker
Evaluate a password against simple strength rules.

```python
"""Exercise 6: Check password strength against a few rules."""
password = input("Enter a password to check: ")

missing_requirements = []
if len(password) < 8:
    missing_requirements.append("be at least 8 characters long")
if not any(char.isupper() for char in password):
    missing_requirements.append("include at least one uppercase letter")
if not any(char.islower() for char in password):
    missing_requirements.append("include at least one lowercase letter")
special_characters = {"@", "#", "$", "%", "&", "*", "!", "?"}
if not any(char in special_characters for char in password):
    missing_requirements.append("include at least one special character (@ # $ % & * ! ?)")

if missing_requirements:
    print("Password must:")
    for requirement in missing_requirements:
        print(f"- {requirement}")
else:
    print("Password is strong")
```

### Exercise 7: Caesar Cipher
Shift letters in a message forward or backward by a fixed key.

```python
"""Exercise 7: Implement a simple Caesar cipher."""
def caesar_cipher(message, key):
    """Shift each alphabetical character by key positions."""
    shifted_characters = []
    for char in message:
        if char.isalpha():
            base = ord("A") if char.isupper() else ord("a")
            offset = (ord(char) - base + key) % 26
            shifted_characters.append(chr(base + offset))
        else:
            shifted_characters.append(char)
    return "".join(shifted_characters)

mode = input("Choose mode (encrypt/decrypt): ").strip().lower()
text = input("Enter the message: ")
key_input = input("Enter the shift key (number): ")

try:
    key = int(key_input)
    if mode == "encrypt":
        print("Encrypted message:", caesar_cipher(text, key))
    elif mode == "decrypt":
        print("Decrypted message:", caesar_cipher(text, -key))
    else:
        print("Unknown mode. Please type 'encrypt' or 'decrypt'.")
except ValueError:
    print("The key must be a number.")
```

### Exercise 8: Hangman Game
Guess letters in a hidden word before running out of attempts.

```python
"""Exercise 8: Play a tiny Hangman game."""
import random

words = ["python", "java", "javascript"]
max_guesses = 6

secret_word = random.choice(words)
display = ["_"] * len(secret_word)
guessed_letters = set()
wrong_guesses = 0

while wrong_guesses < max_guesses and "_" in display:
    print("Word:", " ".join(display))
    guess = input("Guess a letter: ").strip().lower()

    if len(guess) != 1 or not guess.isalpha():
        print("Please enter a single letter.")
        continue
    if guess in guessed_letters:
        print("You already guessed that letter.")
        continue

    guessed_letters.add(guess)
    if guess in secret_word:
        # Reveal all positions where the guessed letter appears.
        for index, letter in enumerate(secret_word):
            if letter == guess:
                display[index] = guess
        print("Good guess!")
    else:
        wrong_guesses += 1
        print(f"Wrong guess. {max_guesses - wrong_guesses} guesses remaining.")

if "_" not in display:
    print(f"Congratulations, you guessed the word '{secret_word}'!")
else:
    print(f"Out of guesses. The word was '{secret_word}'.")
```
